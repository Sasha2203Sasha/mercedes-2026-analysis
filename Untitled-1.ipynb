```python
import fastf1 as ff1
import pandas as pd
import logging

logging.basicConfig(level=logging.INFO)
ff1.Cache.enable_cache('data/')

# 9 этапов, которые точно прошли до 3 июля 2026
RACES_2026_PAST = [
'Australian Grand Prix',
'Chinese Grand Prix',
'Japanese Grand Prix',
'Miami Grand Prix',
'Canadian Grand Prix',
'Monaco Grand Prix',
'Barcelona Grand Prix', # ← бывший Spanish
'Austrian Grand Prix'
# British будет 05.07 — его ещё нет
]

def load_2026_season():
    all_laps = []
    for race_name in RACES_2026_PAST:
        try:
            print(f"🔄 {race_name}...")
            session = ff1.get_session(2026, race_name, 'R')
            session.load()

# Рассел + Антонелли (если код 'ANT' не работает — увидим)
            for code in ['RUS', 'ANT']:
                try:
                    laps = session.laps.pick_driver(code)
                    laps['RaceName'] = race_name
                    laps['Driver'] = code
                    all_laps.append(laps)
                    print(f" ✅ {code}: {len(laps)} кругов")
                except Exception as e:
                    print(f" ⚠️ {code}: {e}")

        except Exception as e:
                print(f"❌ {race_name}: {e}")

                if all_laps:
                    df = pd.concat(all_laps, ignore_index=True)
                    df.to_csv('outputs/mercedes_2026_raw.csv', index=False)
                    print(f"\n🎉 Загружено {len(df)} кругов. Сохранено в outputs/mercedes_2026_raw.csv")
                    return df
                return None

df_raw = load_2026_season()
```

```python
import fastf1 as ff1
import pandas as pd

ff1.Cache.enable_cache('data/')

RACES = [
'Australian Grand Prix', 'Chinese Grand Prix', 'Japanese Grand Prix',
'Miami Grand Prix', 'Canadian Grand Prix', 'Monaco Grand Prix',
'Barcelona Grand Prix', 'Austrian Grand Prix'
]

ferrari_laps = []
for race in RACES:
    try:
        session = ff1.get_session(2026, race, 'R')
        session.load()
        for code in ['HAM', 'LEC']:
            try:
                laps = session.laps.pick_driver(code)
                laps['RaceName'] = race
                laps['Driver'] = code
                laps['Team'] = 'Ferrari'
                ferrari_laps.append(laps)
                print(f" ✅ {race} — {code}: {len(laps)} кругов")
            except:
                print(f" ⚠️ {code} не найден в {race}")
    except Exception as e:
        print(f"❌ {race}: {e}")

df_ferrari_raw = pd.concat(ferrari_laps, ignore_index=True)
df_ferrari_raw['LapTime_sec'] = df_ferrari_raw['LapTime'].dt.total_seconds()
df_ferrari_raw.to_csv('outputs/ferrari_2026_raw.csv', index=False)
print(f"\n✅ Ferrari загружен: {len(df_ferrari_raw)} кругов")
```

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# --- Шаг 1: Загружаем очищенные данные (если df_clean тоже потерялась — из CSV) ---
try:
    df_clean
except NameError:
    df_raw = pd.read_csv('outputs/mercedes_2026_raw.csv')
    df_raw['LapTime'] = pd.to_timedelta(df_raw['LapTime'])
    df_raw['LapTime_sec'] = df_raw['LapTime'].dt.total_seconds()
    df_clean = df_raw[(df_raw['LapTime_sec'] > 55) & (df_raw['LapTime_sec'] < 150)]
    df_clean = df_clean[df_clean['PitInTime'].isna() & df_clean['PitOutTime'].isna()]
    if 'TrackStatus' in df_clean.columns:
        df_clean = df_clean[df_clean['TrackStatus'] == '1']
        df_clean = df_clean[df_clean['LapNumber'] > 1]
    print("✅ df_clean загружен из CSV")

# --- Шаг 2: Считаем деградацию ---
def calc_degradation(stint_laps):
    if len(stint_laps) < 5:
        return None
    X = stint_laps['TyreLife'].values.reshape(-1, 1)
    y = stint_laps['LapTime_sec'].values
    model = LinearRegression().fit(X, y)
    return model.coef_[0]

deg_results = []
for (race, driver, compound), group in df_clean.groupby(['RaceName', 'Driver', 'Compound']):
    rate = calc_degradation(group)
    if rate is not None:
        deg_results.append({
        'Race': race,
        'Driver': driver,
        'Compound': compound,
        'DegradationRate': round(rate, 4),
        'LapsInStint': len(group)
})

deg_df = pd.DataFrame(deg_results)
deg_df.to_csv('outputs/degradation_rates_2026.csv', index=False)
print(f"✅ deg_df создан и сохранён: {len(deg_df)} отрезков")
avg_deg = deg_df.groupby(['Driver', 'Compound'])['DegradationRate'].mean().unstack()

fig, ax = plt.subplots(figsize=(10, 6))
colors = ['#E6002B', '#FFD700', '#FFFFFF']
avg_deg.plot(kind='bar', ax=ax, color=colors[:len(avg_deg.columns)])

ax.set_ylabel('Деградация (сек/круг)', fontsize=12)
ax.set_title('Деградация шин: Рассел vs Антонелли\nMercedes-AMG, 8 этапов 2026',
fontsize=14, fontweight='bold')
ax.legend(title='Состав шин')
ax.grid(axis='y', alpha=0.2)
plt.xticks(rotation=0)

for container in ax.containers:
    ax.bar_label(container, fmt='%.3f', fontsize=9)

plt.tight_layout()
plt.savefig('outputs/degradation_2026.png', dpi=150)
plt.show()

# --- Шаг 4: Вывод ---
rus = deg_df[deg_df['Driver'] == 'RUS']['DegradationRate'].mean()
ant = deg_df[deg_df['Driver'] == 'ANT']['DegradationRate'].mean()
diff = rus - ant
better = "Рассел" if rus < ant else "Антонелли"

print(f"""
═══════════════════════════════════════
ИТОГИ: MERCEDES-AMG 2026 (8 этапов)
═══════════════════════════════════════

Средняя деградация шин (сек/круг):
• Рассел: {rus:.4f}
• Антонелли: {ant:.4f}

Разница: {abs(diff):.4f} сек/круг — лучше бережёт шины {better}

За 20-круговой отрезок это даёт ~{abs(diff)*20:.1f} сек преимущества.

Данные: FastF1 API (первые 8 этапов 2026)
═══════════════════════════════════════
""")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

plt.style.use('dark_background')

# ─── ЗАГРУЗКА ОБЕИХ КОМАНД ───
merc = pd.read_csv('outputs/mercedes_2026_raw.csv')
ferr = pd.read_csv('outputs/ferrari_2026_raw.csv')

merc['Team'] = 'Mercedes'
ferr['Team'] = 'Ferrari'

df_raw = pd.concat([merc, ferr], ignore_index=True)
df_raw['LapTime'] = pd.to_timedelta(df_raw['LapTime'])
df_raw['LapTime_sec'] = df_raw['LapTime'].dt.total_seconds()

# ─── ОЧИСТКА ───
df = df_raw[(df_raw['LapTime_sec'] > 55) & (df_raw['LapTime_sec'] < 150)]
df = df[df['LapNumber'] > 1]
df = df[df['TrackStatus'] == 1]

print(f"✅ Кругов: {len(df)}")
print(f"👤 Пилоты: {df['Driver'].unique()}")
print(f"🏎️ Команды: {df['Team'].unique()}")

# ─── ГРАФИК 1: Бокс-плот темпа (4 пилота) ───
drivers_order = ['RUS', 'ANT', 'HAM', 'LEC']
labels = ['Russell\n(Mercedes)', 'Antonelli\n(Mercedes)', 'Hamilton\n(Ferrari)', 'Leclerc\n(Ferrari)']
colors = ['#00D2BE', '#00D2BE', '#DC0000', '#DC0000']

data = [df[df['Driver'] == d]['LapTime_sec'].dropna().values for d in drivers_order]

fig, ax = plt.subplots(figsize=(14, 6))
bp = ax.boxplot(data, patch_artist=True)
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.85)
    for el in bp['medians']:
        el.set_color('white')
        el.set_linewidth(2)

ax.set_xticklabels(labels, fontsize=10)
ax.set_ylabel('Время круга (сек)', fontsize=12)
ax.set_title('Гоночный темп: Mercedes vs Ferrari\nПервые 8 этапов 2026', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.2)

for i, d in enumerate(data):
    med = np.median(d)
    ax.text(i+1, med+0.4, f'{med:.1f}с', ha='center', fontsize=10, color='white')

plt.tight_layout()
plt.savefig('outputs/pace_4drivers.png', dpi=150)
plt.show()

# ─── ГРАФИК 2: Деградация по пилотам ───
def calc_deg(stint):
    if len(stint) < 5: 
        return None
    X = stint['TyreLife'].values.reshape(-1, 1)
    y = stint['LapTime_sec'].values
    return LinearRegression().fit(X, y).coef_[0]

results = []
for (driver, compound), grp in df.groupby(['Driver', 'Compound']):
    team = grp['Team'].iloc[0]
    r = calc_deg(grp)
    if r:
        results.append({'Driver': driver, 'Team': team, 'Compound': compound,
        'DegradationRate': round(r, 4), 'Laps': len(grp)})
    deg_df = pd.DataFrame(results)

# График по пилотам
avg_driver = deg_df.groupby(['Driver', 'Compound'])['DegradationRate'].mean().unstack()
fig, ax = plt.subplots(figsize=(12, 6))
avg_driver.loc[drivers_order].plot(kind='bar', ax=ax, color=['#E6002B', '#FFD700', '#FFFFFF'])
ax.set_ylabel('Деградация (сек/круг)', fontsize=12)
ax.set_title('Деградация шин по пилотам: Mercedes vs Ferrari\n8 этапов 2026', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.2)
plt.xticks(rotation=0)
for c in ax.containers:
    ax.bar_label(c, fmt='%.3f', fontsize=8)
    plt.tight_layout()
    plt.savefig('outputs/deg_4drivers.png', dpi=150)
    plt.show()

# ─── ГРАФИК 3: Деградация по командам ───
avg_team = deg_df.groupby(['Team', 'Compound'])['DegradationRate'].mean().unstack()
fig, ax = plt.subplots(figsize=(8, 6))
avg_team.plot(kind='bar', ax=ax, color=['#E6002B', '#FFD700', '#FFFFFF'])
ax.set_ylabel('Деградация (сек/круг)', fontsize=12)
ax.set_title('Деградация шин: Mercedes vs Ferrari (среднее)\n8 этапов 2026', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.2)
plt.xticks(rotation=0)
for c in ax.containers:
    ax.bar_label(c, fmt='%.3f', fontsize=10)
    plt.tight_layout()
    plt.savefig('outputs/deg_teams.png', dpi=150)
    plt.show()

# ─── ВЫВОД ───
merc_deg = deg_df[deg_df['Team'] == 'Mercedes']['DegradationRate'].mean()
ferr_deg = deg_df[deg_df['Team'] == 'Ferrari']['DegradationRate'].mean()
better_team = "Mercedes" if merc_deg < ferr_deg else "Ferrari"

rus = deg_df[deg_df['Driver'] == 'RUS']['DegradationRate'].mean()
ant = deg_df[deg_df['Driver'] == 'ANT']['DegradationRate'].mean()
ham = deg_df[deg_df['Driver'] == 'HAM']['DegradationRate'].mean()
lec = deg_df[deg_df['Driver'] == 'LEC']['DegradationRate'].mean()

best_driver = min([('Рассел', rus), ('Антонелли', ant), ('Хэмилтон', ham), ('Леклер', lec)], key=lambda x: x[1])

print(f"""
═══════════════════════════════════════
ИТОГ: MERCEDES vs FERRARI (8 гонок)
═══════════════════════════════════════

Средняя деградация шин (сек/круг):
Mercedes:
• Рассел: {rus:.4f}
• Антонелли: {ant:.4f}
Ferrari:
• Хэмилтон: {ham:.4f}
• Леклер: {lec:.4f}

По командам:
• Mercedes: {merc_deg:.4f} сек/круг
• Ferrari: {ferr_deg:.4f} сек/круг
• Разница: {abs(merc_deg-ferr_deg):.4f} сек/круг

Лучше работают с шинами: {better_team}
Лучший пилот по деградации: {best_driver[0]} ({best_driver[1]:.4f})

За 20 кругов {better_team} выигрывает ~{abs(merc_deg-ferr_deg)*20:.1f} сек.
═══════════════════════════════════════
""")

# Сохраняем
df.to_csv('outputs/combined_2026_clean.csv', index=False)
deg_df.to_csv('outputs/degradation_all_2026.csv', index=False)
print("✅ Сохранено: combined_2026_clean.csv, degradation_all_2026.csv")
```
